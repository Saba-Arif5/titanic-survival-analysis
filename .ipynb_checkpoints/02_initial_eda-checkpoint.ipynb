```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("../datasets/Titanic-Dataset.csv")
```

## Step 1: Count of Survivors by Class

- calculating avg survival per group **survived in each class** (`Pclass`).  
 **Evaluation:**  
- Did higher-class passengers survive more often?


```python
survival_rates= df.groupby('Pclass')['Survived'].mean()
print(survival_rates)                         
```

##  Step 2: Bar Chart of Survivors by Class

- Visualizing the data using a **bar chart** to see how survival varied by class.



```python
# Plot the bar chart
plt.figure(figsize=(6, 4))
plt.bar(survival_rates.index,survival_rates.values, color='skyblue')
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survivors by Passenger Class")
plt.xticks([1, 2, 3])  # Ensuring correct class labels
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Step 3: Line Chart for Comparison
- Trying plotting the same data using a **line chart** â€” it's not the best for categorical data but shows trends.

 **Evaluation:**  
- Does the trend line suggest that survival dropped as class number increased?


```python
# Line plot (for trend)
plt.figure(figsize=(6, 4))
plt.plot(survival_rates.index, survival_rates.values, marker='o', linestyle='-', color='green')
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Trend of Survival by Class (Line Chart)")
plt.xticks([1, 2, 3])
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Observations

From the bar chart and survival rates, I observe that:

-  **1st Class passengers had the highest number of survivors**
-  Survival decreased with class â€” possibly reflecting access to lifeboats or social privilege
