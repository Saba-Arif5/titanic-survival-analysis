### Data Visualization + EDA Notebook

#### Summary:
In this notebook, I practiced different data visualization techniques using Matplotlib and Seaborn on the Titanic dataset. The goal was to understand how visual patterns reveal relationships between variables such as survival rate, age, fare, and passenger class. This included bar charts, histograms, KDE plots, boxplots, and a correlation heatmap.These visualizations collectively strengthened my exploratory data analysis skills and helped interpret trends in the Titanic dataset more effectively.


**Goal:** To learn and apply essential data visualization techniques across Matplotlib and Seaborn by generating meaningful Titanic dataset plots. The objective is to analyze relationships between key variables, identify patterns in survival outcomes, and build strong EDA foundations for future modeling work.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/Titanic-Dataset.csv")
```

```python
survival_rates= df.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(8, 5))  # Wider figure

bars = plt.bar(
    x=survival_rates.index,
    height=survival_rates.values,
    
    color="cyan",  # Professional palette
    edgecolor='black',  # Add borders
    linewidth=1.4       # Border thickness
)
plt.title("Titanic Survival Rates by Passenger class", fontsize=16, pad=10, fontweight='bold')
plt.xlabel("Passenger Class", fontsize=12, labelpad= 8)
plt.ylabel("Survival Rate (%)",fontsize=12, labelpad= 8)
# --- Axis Customization ---
plt.xticks([1, 2, 3], labels=["First Class", "Second Class", "Third Class"], fontsize=12)
plt.yticks(fontsize=12)  # Adjust y-axis font
plt.ylim(0, 0.7)  # Set y-axis range
# --- Grid & Spines ---
ax = plt.gca()
ax.yaxis.grid(True, linestyle='--', alpha=0.4)  # More explicit
plt.gca().spines['top'].set_visible(False)    # Remove top border
plt.gca().spines['right'].set_visible(False)  # Remove right border
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0%}',  # Show % without decimals
             ha='center', va='bottom', fontsize=12)

plt.tight_layout()  # Prevent label clipping
plt.show()

```

### Titanic Survival Rates by Passenger Class

### What I Did
- Grouped passengers by `Pclass` and calculated mean survival rates.  
- Created a styled bar chart using Matplotlib with colors, borders, custom labels, and percentage annotations.

### What the Plot Shows
- The chart displays survival rates for First, Second, and Third Class.  
- Higher bars indicate higher survival probability by class.  
- First Class shows the highest rate, followed by Second and then Third.

### Insights
Survival clearly depended on passenger class, with First Class being significantly more likely to survive. This supports the historical pattern that socio-economic status influenced evacuation priority on the Titanic.


# Histograms: Age Distribution by Survival Status (Seaborn)-learning basics of seaborn d10

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("datasets/Titanic-Dataset.csv")
```

```python
# Set style
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x="Age",
    hue="Survived",
    bins=20,
    stat="probability",
    common_norm=False,
    #kde=True,
    palette={0:"Red", 1:"blue"},
    alpha=0.6)
plt.title("Age Distribution by Survival Status", pad=20)
plt.xlabel("Age(years)")
plt.ylabel("probability")
plt.legend(title="Survived", labels=["Yes", "No"])
# # Add vertical lines for key age thresholds
# plt.axvline(x=18, color='blue', linestyle='--', alpha=0.3, label='Child/Adult Threshold')
# plt.axvline(x=60, color='purple', linestyle='--', alpha=0.3, label='Senior Threshold')
# # Add text annotation for children's high survival
# plt.text(x=10, y=0.15, 
#          s="Children (<18) had\nhigher survival rates", 
#          ha='center', fontsize=10,
#          bbox=dict(facecolor='white', alpha=0.8))
# plt.show()
```

### KDE + Histogram for Passenger Age

### What I Did
- Plotted a Seaborn histogram with KDE enabled to visualize age density more smoothly.  
- Added `hue="Survived"` to compare survivors and non-survivors.

### What the Plot Shows
- KDE curves reveal the underlying age distribution for both groups.  
- The shape shows peaks around childhood and early adulthood.

### Insights
The KDE curves confirm the earlier observation: while survival varies slightly across age, both groups share similar overall age patterns. This reinforces that age affects survival only moderately.

#### Box Plot and Heat Maps

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("datasets/Titanic-Dataset.csv")

```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Pclass', y='Fare', data= df)
plt.title("Fare distribution by Passenger class")
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

### Fare Distribution by Passenger Class (Boxplot)

#### What I Did
- Created a boxplot comparing `Fare` across the three passenger classes.  
- Used `Pclass` on the x-axis and `Fare` on the y-axis to visualize distribution differences.

#### What the Plot Shows
- First-class passengers have the highest median fares.  
- Third-class fares are the lowest with less variation.  
- Presence of outliers shows some very expensive tickets, mostly in first class.

#### Insights
Fare clearly increases with passenger class, confirming strong socioeconomic differences on the Titanic.



```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```


### Age Distribution by Survival (Boxplot)

#### What I Did
- Plotted a boxplot with `Survived` as the category and `Age` as the numerical distribution.  
- Compared age ranges of survivors vs. non-survivors.

#### What the Plot Shows
- Survivors tend to have slightly lower median ages.  
- Both groups contain a wide age range, including children and adults.  
- Outliers indicate a few unusually old or young passengers.

#### Insights
Younger passengers generally had better survival chances, though age alone does not fully explain the outcome.


```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Titanic Data')
plt.show()
```

### Correlation Heatmap of Titanic Dataset

### What I Did
- Generated a correlation heatmap using Seaborn with numeric columns only.  
- Displayed correlation values using color intensity and annotations.

### What the Plot Shows
- Strong negative correlation between `Pclass` and `Fare`.  
- Moderate positive correlation between `Fare` and `Survived`.  
- Weak correlations for most other features.

### Insights
Fare and passenger class show the strongest relationship to survival among numeric variables. This highlights socio-economic status as a key survival factor.


## Side-by-Side Histogram

```python
fig, axes=plt.subplots(1,2,figsize=(12,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, ax=axes[0])
axes[0].set_title("Age Distribution of Passengers")

sns.histplot(df['Fare'].dropna(), bins=30, kde=True, ax=axes[1])
axes[1].set_title("Fare Distribution of Passengers")

plt.tight_layout()
plt.show()
```

### Age and Fare Distributions (Side-by-Side Histograms)

#### What I Did
- Created a two-panel figure using `plt.subplots()` to compare distributions side by side.  
- Plotted a histogram with KDE for `Age` on the left.  
- Plotted a histogram with KDE for `Fare` on the right.  
- Used `.dropna()` to avoid errors from missing values.

#### What the Plots Show
- **Age Distribution:**  
  - Most passengers were between 2040 years old.  
  - KDE curve shows a smooth peak around young adulthood.  
- **Fare Distribution:**  
  - Fare values are heavily right-skewed (most fares low, few very expensive).  
  - KDE shows a long tail due to high-fare outliers.

#### Quick Interpretation
The majority of Titanic passengers were young adults, while fare prices varied widely with many low-cost tickets and a small number of high-fare outliers. This pattern reflects the socioeconomic diversity of the passengers.


### Evaluation Prompts
1. Looking at both plots, do age and fare show similar or very different patterns of distribution?

They show **very different** patterns. Age is roughly bell-shaped with most passengers between 2040, while Fare is heavily skewed toward low values with a few extremely high fares.

2. Are there many outliers in the fare distribution? What might they represent?

Yes, there are clear outliers on the high end of the fare axis. These likely represent **first-class or luxury cabin passengers** who paid significantly more than most others.

3. How does this side-by-side view make analysis easier compared to plotting separately?

It allows a **direct visual comparison** without switching between separate figures. Seeing them together makes differences in distribution shape and skewness obvious at a glance.

   


### Comparing Survival by Sex vs. Class

We want to see survival rates broken down by **sex** and by **class**, side by side.  
This lets us compare two important demographic factors in one view.

---


```python


# Set up side-by-side plots
fig, axes =plt.subplots(1,2, figsize=(12,5))
#Survival rate by sex
sns.barplot(x="Sex", y="Survived", data=df, ax=axes[0])
axes[0].set_title("Survived by gender")
axes[0].set_ylabel("Survival Rate")

#Survival rate by passenger class
sns.barplot(x="Pclass", y="Survived",data=df, ax=axes[1])
axes[1].set_title("Survival rate by class")
axes[1].set_ylabel("Survival Rate")

plt.tight_layout()
plt.show()
```

### Survival Rates by Gender and Passenger Class (Side-by-Side Barplots)

#### What I Did
- Used `plt.subplots()` to create two bar charts side by side.  
- First plot shows average survival rate grouped by gender.  
- Second plot shows average survival rate grouped by passenger class.

#### What the Plots Show
- **Gender Plot:**  
  - Females have a significantly higher survival rate than males.  
- **Class Plot:**  
  - First-class passengers have the highest survival rate.  
  - Survival decreases progressively in second and third class.

#### Insights
Gender and passenger class both had strong influence on survival: women and higher-class passengers were much more likely to survive, aligning with historical accounts of evacuation priorities.
